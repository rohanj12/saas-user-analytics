```python
#Convert this column to a numeric one
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')

#Remove all null values
df.dropna(subset=['TotalCharges'], inplace=True)

#Reset the index
df.reset_index(drop=True, inplace=True)
```

## 2. EXPLORATORY DATA ANALYSIS

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv('/Users/rohanjairam/projects/saas-user-analytics/data/raw_data.csv')
```

```python
churn_counts = df['Churn'].value_counts()
churn_rate = df['Churn'].value_counts(normalize=True) * 100

print("Churn Rate (%):\n", churn_rate)
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', kde=True, bins=30)
plt.title("Tenure Distribution by Churn Status")
plt.xlabel("Months with Company")
```

```python
numerics = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numerics.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Churn', y='TotalCharges', data=df)
plt.title("Total Charges by Churn Status")
```
